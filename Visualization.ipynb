```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Exp -> 1 Queue 1 Frame

```python
data = pd.read_csv('log/1frame_constant.csv', header=None)

plt.figure(figsize=(12,8))
sns.lineplot(range(len(data[0])), data[0], color='gray')
plt.xlabel('Epoch')
plt.ylabel('Reward')
plt.savefig('1.pdf')

plt.figure(figsize=(12,8))
sns.lineplot(range(len(data[1])), data[1], color='gray')
plt.xlabel('Epoch')
plt.ylabel('Queue length')
plt.savefig('2.pdf')
```

```python
data = pd.read_csv('log/1frame_dynamic.csv', header=None)

plt.figure(figsize=(12,8))
sns.lineplot(range(len(data[0])), data[0], color='gray')
plt.xlabel('Epoch')
plt.ylabel('Reward')
plt.savefig('3.pdf')

plt.figure(figsize=(12,8))
sns.lineplot(range(len(data[1])), data[1], color='gray')
plt.xlabel('Epoch')
plt.ylabel('Queue length')
plt.savefig('4.pdf')
```
